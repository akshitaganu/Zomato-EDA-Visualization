```python
#removing col url, address, phone, menu item, dish linked, reviews list
```

```python
df = df.drop('phone', axis=1)
```

```python
df = df.drop('menu_item', axis=1)
```

```python
df
```

```python
df.info()
```

# Dropping dupplicates

```python
df.drop_duplicates(inplace = True)
df.shape
```

# Cleaning rate col

```python
df['rate'].unique()
```

```python
#now i want to remove the rows which has 'NEW', '-', 'nan' and /5 values 
```

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

```python
#filling the null values in rate col by replacing mean of that rate col
```

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

# Dropping null values

```python
df.dropna(inplace = True)
```

```python
df.head()
```

```python
#shorter name for col
df.rename(columns = {'approx_cost(for two people)': 'Cost2plates', 'listed_in(type)': 'Type'}, inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
#listed in city and location both are there, so we will keep only one
df = df.drop(['listed_in(city)'], axis = 1)
```

```python
df.head()
```

```python
df['Cost2plates'].unique()
```

```python
#remov coma from every value

def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
        
```

```python
df.head()
```

# Cleaning Rest type col

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
# Making Rest Types less than 1000 in frequency as others
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

# cleaning location col

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending  = False)
location
```

```python
location_lessthan300 = location[location<300]
location_lessthan300
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

# cleaning cuisines col

```python
cuisines = df['cuisines'].value_counts(ascending  = False)
cuisines_lessthan100 = cuisines[cuisines<100]


def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
#since type col has very less unique values...we cant cluster them 

df['Type'].value_counts()
```
